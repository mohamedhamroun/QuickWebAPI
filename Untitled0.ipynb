cleaning dataset


```python
#nltk.download("stopwords")
#nltk.download('wordnet')

def cleaning(df, stop_words):

  df['tweet_text'] = df['tweet_text'].apply(lambda x:  ' '.join(x.lower() for x in x.split()))
 # Replacing the special characters

  #df['tweet_text'] = re.sub(r"[^a-zA-Z0-9]+", ' ', df['tweet_text'])

 # Replacing the digits/numbers
  df['tweet_text'] = df['tweet_text'].str.replace('d', '' )

 # Removing stop words
  df['tweet_text'] = df['tweet_text'].apply(lambda x: ' '.join(x for x in x.split() if x not in stop_words))

 # Lemmatization
  df['tweet_text'] = df['tweet_text'].apply(lambda x:  ' '.join([Word(x).lemmatize() for x in x.split()]))

  return df


#stop_words = stopwords.words('english')

#data_v1 = cleaning(data, stop_words)
```
